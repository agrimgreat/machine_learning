### Q2 Calculate the Variance Inflation Factor (VIF) value. VIF is a number that determines whether a variable has multicollinearity or not (starts from 1, and it has no upper limit. If the number gets larger, it means the variable has huge multicollinearity on it.)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
from statsmodels.tools.tools import add_constant

numeric_df=df.select_dtypes(include=['number'])
X = add_constant(numeric_df)
for i in range(X.shape[1]):
    print(X.columns[i],'\t',variance_inflation_factor(X.values, i),'\n')
```

### Q3Remove multicollinearities by creating new features. Find the features that have paired values and create the new feature which is the difference value between those pairs.

```python
df['temp_diff']=df['MaxTemp']-df['MinTemp']
df['humidity_diff']=df['Humidity9am']-df['Humidity3pm']
df['temp_time_diff']=df['Temp9am']-df['Temp3pm']

df2=df.drop(columns=['MaxTemp','MinTemp','Humidity9am','Humidity3pm','Temp9am','Temp3pm'])
```

### Q4 Remove features that have a VIF value above 5

```python
numeric_df2=df2.select_dtypes(include=['number'])
X2=add_constant(numeric_df2)
for i in range(X2.shape[1]):
    print(X2.columns[i],'\t',variance_inflation_factor(X2.values, i),'\n')
```

### Q5 Build a regression model to perform the Rain prediction. Also, tabulate accuracy of the prediction models, before and VIF computation.


```python
df=df.drop(columns=['temp_diff','humidity_diff','temp_time_diff'])
print(df.dtypes,'\n')
print(df2.dtypes)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

X_before_vif=df.select_dtypes(include=['number'])
y_before_vif=df['RainTomorrow']

X_train_before_vif,X_test_before_vif,y_train_before_vif,y_test_before_vif=train_test_split(X_before_vif,y_before_vif,test_size=0.2, random_state=42)

model=LogisticRegression(max_iter=1000)
model.fit(X_train_before_vif,y_train_before_vif)

y_pred_before_vif=model.predict(X_test_before_vif)


accuracy=accuracy_score(y_test_before_vif,y_pred_before_vif)
accuracy
```

```python
X_after_vif=df2.select_dtypes(include=['number'])
y_after_vif=df2['RainTomorrow']

X_train_after_vif,X_test_after_vif,y_train_after_vif,y_test_after_vif=train_test_split(X_after_vif,y_after_vif,test_size=0.2, random_state=42)

model=LogisticRegression(max_iter=1000)
model.fit(X_train_after_vif,y_train_after_vif)

y_pred_after_vif=model.predict(X_test_after_vif)


accuracy=accuracy_score(y_test_after_vif,y_pred_after_vif)
accuracy
```
